#### Let's change the pattern of this dataset, to predicting today's rainfall using bistorical weather data up to and including yesterday, then we can legitimately utilize all of the available features. 
With this, we can decide if we are going to take an umbrella today or not. 


```python
## Update the names of the rain columns to avoid confusion. 

df = df.rename(columns={'RainToday': 'RainYesterday',
                        'RainTomorrow': 'RainToday'
                        })
```

### Location Selection

The chance of it rainining in one location in Australia can be higher than the other, After grouping cities in he Location column by distance, we found out that Watsonia is only 15 km from Melbourne, and the Melbourne Airport is only 18 km from Melbourne.

We will group the three lcoations together and use only their weather data to build our localized prediction model.

```python
df = df[df.Location.isin(['Melbourne','MelbourneAirport','Watsonia',])]
df. info()
```

Now consider the Date column. We expect the weather patterns to be seasonal, having different predictablitiy levels in winter and summer for example.

#### Map dates to Seasons

```python
def date_to_season(date):
    month = date.month
    if (month == 12) or (month == 1) or (month == 2):
        return 'Summer'
    elif (month == 3) or (month == 4) or (month == 5):
        return 'Autumn'
    elif (month == 6) or (month == 7) or (month == 8):
        return 'Winter'
    elif (month == 9) or (month == 10) or (month == 11):
        return 'Spring'
```

```python
# Map dates to seasons and drop the Date Column
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Apply the function to the "Date' Column
df['Season'] = df['Date'].apply(date_to_season)

df = df.drop(columns='Date')
df
```

### Let's define the feature and the target dataframes

```python
# Write your response.
X = df.drop(columns='RainToday', axis=1)
y= df['RainToday']
```

```python
## Check the balance of the target variable 
y.value_counts()
```

From the above count, I discoverd that \
1. It rains around 86 days annually thats around 24%.
2. I am be 76% sure that it won't rain.
3. This is not a balanced dataset.

#### We split the datset to trianing and test sets and stratify because of the imbalance ration of the target variable

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify = y, random_state =42)
```

```python
## Detect the numeric and categorical columns

numeric_features = X_train.select_dtypes(include=['number']).columns.tolist()
categorical_features = X_train.select_dtypes(include=['object', 'category']).columns.tolist()
```

```python
## Define the transformers for both feature types
# Scale numeric features
numeric_transformer = Pipeline(steps = [('scaler', StandardScaler())])

# Encode categorical features
categorical_transformer = Pipeline(steps= [('onehot', OneHotEncoder(handle_unknown='ignore'))])
```

```python
## Combine the transformers into a single ColumnTransformer

preprocessor = ColumnTransformer(
    transformers = [
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ]
)
```

```python
### Create a pipeline for the model

pipeline = Pipeline(steps = [
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42   ))
])
```

```python
## define parameter grid in cross validation grid search
param_grid = {
    'classifier__n_estimators': [50, 100],
    'classifier__max_depth': [None, 10, 20],
    'classifier__min_samples_split': [2, 5]
}
```

```python
## Select a cross-validation method, ensuring target stratification during validation
cv = StratifiedKFold(n_splits=5, shuffle=True)
```

```python
### fit the GridSearchCV to the pipeline

grid_search = GridSearchCV(pipeline, param_grid, cv=cv, scoring='accuracy', verbose=2)
grid_search.fit(X_train, y_train)
```

```python
## print the best parameters and best cross validation score
print("\nBest parameters found: ", grid_search.best_params_)
print("Best cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
## Display the model estimated score
test_score = grid_search.score(X_test, y_test)
print("Test set score: {:.2f}".format(test_score))
```

So we have a reasonably accurate classifer, which is expected to correctly predict about 84% of the time whether it will rain today in the Melbourne area.

#### Get the model predictions form the grid search estimator on the unseen data

```python
y_pred = grid_search.predict(X_test)
```

```python
#print classification report

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
## Plot the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix = conf_matrix)
disp.plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

The true positive is 183, the false positive is 175. The true false is 1090 and the false false is 64

## Extract the Feature importance

```python
feature_importances = grid_search.best_estimator_['classifier'].feature_importances_
```

```python
## extract the feature importance and plot them

# Combine numeric and categorical feature names
feature_names = numeric_features + list(grid_search.best_estimator_['preprocessor']
                                        .named_transformers_['cat']
                                        .named_steps['onehot']
                                        .get_feature_names_out(categorical_features))

feature_importances = grid_search.best_estimator_['classifier'].feature_importances_

importance_df = pd.DataFrame({'Feature': feature_names,
                              'Importance': feature_importances
                             }).sort_values(by='Importance', ascending=False)

N = 20  # Change this number to display more or fewer features
top_features = importance_df.head(N)

# Plotting
plt.figure(figsize=(10, 6))
plt.barh(top_features['Feature'], top_features['Importance'], color='skyblue')
plt.gca().invert_yaxis()  # Invert y-axis to show the most important feature on top
plt.title(f'Top {N} Most Important Features in predicting whether it will rain today')
plt.xlabel('Importance Score')
plt.show()
```

The humidity3pm is a great feature that strongly determines if it is going to rain.
